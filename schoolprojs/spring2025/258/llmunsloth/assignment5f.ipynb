```python
!pip install transformers datasets
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments, DataCollatorForLanguageModeling
from datasets import Dataset
examples = [
    {
        "text": "### Instruction: I feel anxious all the time. What should I do?\n### Response: It's okay to feel anxious. Try breathing exercises, talk to someone you trust, or consult a mental health professional."
    },
    {
        "text": "### Instruction: I can't sleep due to stress.\n### Response: Try a relaxing routine before bed, avoid screens, and practice mindfulness or journaling."
    },
    {
        "text": "### Instruction: I'm afraid to talk to people because of social anxiety.\n### Response: You're not alone. Start small, be kind to yourself, and seek support if you need it."
    }
]

dataset = Dataset.from_list(examples)

def tokenize(example):
    tokens = tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)
    tokens["labels"] = tokens["input_ids"]
    return tokens
```

```python
model_id = "sshleifer/tiny-gpt2"  # small, CPU-friendly
tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_id)

tokenized_data = dataset.map(tokenize)
training_args = TrainingArguments(
    output_dir="./mh-chatbot",
    per_device_train_batch_size=1,
    num_train_epochs=1,
    logging_steps=1,
    save_strategy="no",
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_data,
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False),
)

trainer.train()
```

```python
from transformers import pipeline

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

prompt = "### Instruction: I've been feeling really down lately.\n### Response:"
response = pipe(prompt, max_new_tokens=50)[0]['generated_text']

print("🧠 Chatbot Response:\n", response)
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

model_id = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

prompt = "### Instruction: I feel really anxious and can't sleep well.\n### Response:"
response = pipe(prompt, max_new_tokens=60)[0]['generated_text']

print(response)
```
