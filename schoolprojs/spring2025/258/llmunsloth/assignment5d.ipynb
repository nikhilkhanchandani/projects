```python
# Install required libraries

!pip install --upgrade --quiet torch torchvision torchaudio
!pip install --quiet transformers

# Load GPT-2 model and tokenizer
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

preference_examples = [
    {
        "prompt": "What is the capital of France?",
        "chosen": "The capital of France is Paris.",
        "rejected": "France is the name of a planet."
    },
    {
        "prompt": "How do I make scrambled eggs?",
        "chosen": "To make scrambled eggs, beat eggs in a bowl and cook them on a pan with butter.",
        "rejected": "Scrambled eggs are made from boiling water and sugar."
    }
]

# Display them
for ex in preference_examples:
    print("Prompt:", ex["prompt"])
    print("Chosen (better):", ex["chosen"])
    print("Rejected (worse):", ex["rejected"])
    print("-" * 50)
# Try generating based on one of the prompts
prompt = "What is the capital of France?\nAnswer:"
output = generator(prompt, max_new_tokens=50, do_sample=True)[0]["generated_text"]
print("GPT-2 Output:\n", output)
```
